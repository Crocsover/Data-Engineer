```python
import pandas as pd


sales = pd.read_csv(r'C:\Users\public\sales.csv')


sales.head()
```

```python
sales = pd.DataFrame(sales)
print(sales)
```

```python
drop_df = sales.drop_duplicates()
print(drop_df)
```

```python


# Convert the "Price" column to numeric (float) data type
sales['Price'] = pd.to_numeric(sales['Price'], errors='coerce')

# Perform the division to express it in thousands
sales['Price'] = sales['Price'] / 1000

# Print the updated DataFrame to verify the changes
print(sales)
```

```python
#  filter condition
filter_condition = (sales['Product'] == 'Product1') & \
                   (sales['Country'] == 'United States') & \
                   ((sales['Payment_Type'] == 'Amex') | (sales['Payment_Type'] == 'Diners'))

# Apply the filter condition to the DataFrame and select specific columns
selected_data = sales.loc[filter_condition, ['Name', 'City','State', 'Country']]

# Print the selected data
print(selected_data)
```

```python
output_file_path = 'C:\\Users\\public\\empty.csv'


selected_data.to_csv(output_file_path, index=False)


print(f'Results saved to {output_file_path}')
```

```python
#  filter condition
filter_condition1 = (sales['Product'] == 'Product2') & \
                    (sales['City'] == 'New York') & \
                    (sales['Payment_Type'] == 'Mastercard')

# Apply the filter condition to the DataFrame and select specific columns
selected_data1 = sales.loc[filter_condition1, ['Name', 'Price','City','State', 'Country']]

# Print the selected data
print(selected_data1)
```

```python
output_file_path = 'C:\\Users\\public\\Empty.csv'  # Specify your desired file path

# Save the cleaned data to a CSV file
sales.to_csv(output_file_path, index=False)

# Print a message to confirm that the file has been saved
print(f'Cleaned data saved to {output_file_path}')
```
